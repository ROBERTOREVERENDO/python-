## PARTE 19.1: DISTRIBUCIÓN UNIFORME

```python
import numpy as np
import matplotlib.pyplot as plt

#DEFINICIÓN DE LÍMITES
LIM_INF = 40
LIM_SUP = 100

#CANTIDAD DE ESCENARIOS
ESCENARIOS = 1000000

DATASET = np.random.uniform(LIM_INF, LIM_SUP , ESCENARIOS)

plt.hist(DATASET)
plt.title("DISTRIBUCIÓN UNIFORME")
```

#### MEDIA Y DESVIACIÓN POR DEFECTO

```python
import numpy as np
import matplotlib.pyplot as plt

ESCENARIOS = 1000000

DATASET = np.random.randn(ESCENARIOS)

print("MEDIA -->",DATASET.mean())
print("DESVIACIÓN -->",DATASET.std())

plt.hist(DATASET , bins = 200)
plt.title("DISTRIBUCIÓN NORMAL")
```

#### MEDIA Y DESVIACIÓN INGRESADAS

```python
import numpy as np
import matplotlib.pyplot as plt

ESCENARIOS = 1000000
MEDIA = 40
DESVIACION = 30

DATASET = np.random.normal(MEDIA , DESVIACION , ESCENARIOS)

print("MEDIA -->",DATASET.mean())
print("DESVIACIÓN -->",DATASET.std())

plt.hist(DATASET , bins = 200)
plt.title("DISTRIBUCIÓN NORMAL")
```

## PARTE 19.3: DATASET CON DISTRIBUCIONES ESTADÍSTICAS

```python
import numpy as np
import pandas as pd

DEPARTAMENTOS = ["LIMA","ICA","AMAZONAS","TUMBES","CUSCO"]
SEXO = ["MASCULINO","FEMENINO"]
PROFESIONES = ["INGENIERO","DOCTOR","ABOGADO","PSIQUIATRA","PILOTO"]

EDADES = np.random.normal(50, 10 , 1000)
SUELDO = np.random.uniform(1000 , 20000 , 1000)

#GENERACIÓN DE LISTAS VACÍAS
LISTA_DEPARTAMENTO = []
LISTA_SEXO = []
LISTA_PROFESION = []
LISTA_EDAD = []
LISTA_SUELDO = []

for X in range(1000):
    LISTA_DEPARTAMENTO.append(np.random.choice(DEPARTAMENTOS))
    LISTA_SEXO.append(np.random.choice(SEXO))
    LISTA_PROFESION.append(np.random.choice(PROFESIONES))
    LISTA_EDAD.append(int(EDADES[X]))
    LISTA_SUELDO.append(int(SUELDO[X]))
    
DATASET = pd.DataFrame({
    "DEPARTAMENTO":LISTA_DEPARTAMENTO,
    "SEXO":LISTA_SEXO,
    "PROFESIÓN":LISTA_PROFESION,
    "EDAD":LISTA_EDAD,
    "SUELDO":LISTA_SUELDO
})
```
