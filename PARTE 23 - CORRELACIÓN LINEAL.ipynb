## PARTE 23.2: OBTENCIÓN DE COEFICIENTE PEARSON 

```python
import pandas as pd
RUTA = "DATASETS - CERTIFICACIÓN/23_1_CORRELACION_VENTAS.csv"
DATASET = pd.read_csv(RUTA)
DATASET.head()
```

#### OBTENCIÓN DE VALORES MEDIANTE FÓRMULA

```python
import numpy as np
V1 = "TV"
V2 = "Sales"
NUMERADOR = sum((DATASET[V1]-np.mean(DATASET[V1]))*(DATASET[V2]-np.mean(DATASET[V2])))
DENOMINADOR = np.sqrt((sum((DATASET[V1]-np.mean(DATASET[V1]))**2))*(sum((DATASET[V2]-np.mean(DATASET[V2]))**2)))
print("PEARSON -> {}-{} --> {}".format(V1,V2,NUMERADOR/DENOMINADOR))
```

## PARTE 23.3: AUTOMATIZACIÓN DE COEFICIENTE PEARSON

```python
import pandas as pd
RUTA = "DATASETS - CERTIFICACIÓN/23_1_CORRELACION_VENTAS.csv"
DATASET = pd.read_csv(RUTA)
DATASET.head()
```

```python
COLUMNAS = DATASET.columns.tolist()
COLUMNAS
```

```python
for X in COLUMNAS:
    for Y in COLUMNAS:
        if X != Y:
            NUMERADOR = sum((DATASET[X]-np.mean(DATASET[X]))*(DATASET[Y]-np.mean(DATASET[Y])))
            DENOMINADOR = np.sqrt((sum((DATASET[X]-np.mean(DATASET[X]))**2))*(sum((DATASET[Y]-np.mean(DATASET[Y]))**2)))
            print("PEARSON -> {}-{} --> {}".format(X,Y,NUMERADOR/DENOMINADOR))
```
