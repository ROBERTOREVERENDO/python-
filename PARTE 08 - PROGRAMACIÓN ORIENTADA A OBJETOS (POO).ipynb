## PARTE 8.2: EJEMPLO DE PROGRAMACIÓN ESTRUCTURADA

```python
DATOS = [
    {"NOMBRE":"BENJAMIN" , "APELLIDO":"NEIRA" , "EDAD":27},
    {"NOMBRE":"RICARDO" , "APELLIDO":"MORALES" , "EDAD":75},
    {"NOMBRE":"JUAN" , "APELLIDO":"MENDOZA" , "EDAD":23},
    {"NOMBRE":"PIERO" , "APELLIDO":"FALCON" , "EDAD":70}
]
```

#### RECORDAR QUÉ ES ENUMERATE

```python
for INDICE,VALOR in enumerate(DATOS):
    print("INDICE-->",INDICE)
    print("VALOR-->",VALOR)
```

```python
DATOS[0]["NOMBRE"]
```

```python
DATOS[2]["EDAD"]
```

#### LISTAR NOMBRES DE "DATOS"

```python
def LISTAR(LISTA):
    for D in LISTA:
        print(D["NOMBRE"])
        
LISTAR(DATOS)
```

#### BUSCAR NOMBRES EN "DATOS"

```python
def BUSCAR(NOMBRE):
    global DATOS
    
    for D in DATOS:
        if D["NOMBRE"] == NOMBRE:
            print("Sí se encuentra en la lista")
            return
        
    print("No se encuentra en la lista") 
```

```python
BUSCAR("RICARDO")
```

```python
BUSCAR("MANUEL")
```

#### ELIMINAR NOMBRE EN "DATOS"

```python
def ELIMINAR(NOMBRE):
    global DATOS
    
    for INDICE, VALOR in enumerate(DATOS):
        if VALOR["NOMBRE"] == NOMBRE:
            print("Sí se eliminó de la lista")
            DATOS.pop(INDICE)
            return
    print("No se encuentra en la lista") 
```

```python
ELIMINAR("MANUEL")
```

```python
LISTAR(DATOS)
```

```python
ELIMINAR("PIERO")
```

```python
LISTAR(DATOS)
```
