## PARTE 15.1: IMPORTAR UN .CSV MEDIANTE READ_CSV

* VALIDAR QUE SE CARGÓ CORRECTAMENTE
* FILAS --> CONSULTA DE CABECERA Y COLA
* COLUMNAS --> LA ÚLTIMA COLUMNA DEBE TENER INFORMACIÓN CORRECTA

```python
import pandas as pd
RUTA = "DATASETS - CERTIFICACIÓN/15_1_CARGA_CSV_VACUNACION.csv"
DATASET = pd.read_csv(RUTA)
```

#### COMANDOS IMPORTANTES DEL READ_CSV
* names --> NOMBRE DE LAS CABECERAS
* header --> VALIDAR SI TIENE CABECERA O NO
* skiprows --> OBVIAR DETERMINADA CANTIDA DE FILAS
* sep --> INDICAR CUÁL ES EL CARACTER SEPARADOR

```python
CABECERA = ["C1","C2","C3","C4","C5","C6","C7","C8","C9","C10","C11","C12","C13","C14","C15"]

RUTA = "DATASETS - CERTIFICACIÓN/15_1_CARGA_CSV_VACUNACION.csv"
DATASET = pd.read_csv(RUTA,header = None, names = CABECERA, skiprows = 1, sep = ",")
```

## PARTE 15.2: IMPORTAR UN .XLS . XLSX MEDIANTE READ_EXCEL

```python
import pandas as pd
RUTA = "DATASETS - CERTIFICACIÓN/15_2_CARGA_XLS_COMPUTACION_BRASIL.xlsx"
DATASET = pd.read_excel(RUTA,"Sheet1")
```

```python
DATASET.head()
```

#### COMANDOS IMPORTANTES DEL READ_EXCEL
* names --> NOMBRE DE LAS CABECERAS
* header --> VALIDAR SI TIENE CABECERA O NO
* skiprows --> OBVIAR DETERMINADA CANTIDA DE FILAS

```python
CABECERA = ["C1","C2","C3","C4","C5","C6","C7","C8","C9"]

RUTA = "DATASETS - CERTIFICACIÓN/15_2_CARGA_XLS_COMPUTACION_BRASIL.xlsx"
DATASET = pd.read_excel(RUTA,"Sheet1", header = None , names = CABECERA, skiprows = 1)
```

## PARTE 15.3: IMPORTAR UN .TXT MEDIANTE READ_CSV
