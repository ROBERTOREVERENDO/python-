## PARTE 18.14: AGRUPAR UN DATASET MEDIANTE UNA VARIABLE

* count-->Conteo de valores
* sum-->Suma de valores 
* mean-->Promedio de valores 
* median-->Mediana de valores 
* min-->Mínimo valor
* max-->Máximo valor
* mode-->Moda de valores
* prod-->Producto de valores
* std-->Desviación estándar
* var-->Varianza

```python
import pandas as pd
RUTA = "DATASETS - CERTIFICACIÓN/17_1_VISUALIZACION_CHURN_MODEL.csv"
DATASET = pd.read_csv(RUTA)

DATASET.head(10)
```

```python
DATASET_AGRUPADO_STATE = DATASET.groupby(["State"])
DATASET_AGRUPADO_STATE
```

```python
#FILTRAR LA INFORMACIÓN DE UN DATASET AGRUPADO
DATASET_AGRUPADO_STATE.get_group("NJ").head()
```

```python
#AGRUPAR LA INFORMACIÓN CON BASE EN ALGUNA FUNCIÓN
DATASET_AGRUPADO_STATE.mean()
```

```python
#TAMAÑO DE LA CATEGORÍA
DATASET_AGRUPADO_STATE.size()
```

```python
#DESCRIPCIÓN DE LA CATEGORÍA
DATASET_AGRUPADO_STATE.describe()
```

```python
#AGREGADO DE LA INFORMACIÓN
DATASET_AGRUPADO_STATE.agg(
    MAX_Acc_len = ("Account length","max") ,
    STD_Are_cod = ("Area code", "std") ,
    MEAN_NV_Messages = ("Number vmail messages","mean") ,
    VAR_TD_Minutes = ("Total day minutes", "var") ,
    SUM_TD_Calls = ("Total day calls", "sum") ,
    MIN_TD_Charge = ("Total day charge", "min"),
    Acc_len_25 = ("Account length", lambda X : X.quantile(0.25)) ,
    Acc_len_50 = ("Account length", lambda X : X.quantile(0.5)) , 
    Acc_len_75 = ("Account length", lambda X : X.quantile(0.75)) 
)
```

## PARTE 18.17: FILTRAR UN DATASET AGRUPADO

### APOYO PARA ENTENDER ESTE CAPÍTULO:

#### Filtrar en un DATASET agrupado no es lo mismo que filtrar sobre un DATASET sin agrupar 
#### Para entender lo anterior analiza estas 2 preguntas:
1. Dime los registros cuyo "Total day minutes" sea mayor a 185 (filtrado simple)
2. Dime los estados cuyo "Total day minutes" promedio sea mayor a 185 (filtrado sobre agrupado)

#### Para aplicar un filtrado sobre un DATASET agrupado es necesario tener en cuenta que responde a preguntas como la siguiente:
* Dime los registros cuyo "FUNCIÓN AGRUPADORA" de "VARIABLE AGRUPADORA" cumple "CONDICIÓN"

#### Paso a paso a seguir:
1. Implementa una agrupación del DATASET con base a la variable agrupadora
2. Sobre el DATASET agrupado aplica una función filter para aplicar la condición solicitada
3. Obtén los índices de dicha las filas que cumplen dicha condición
4. Filtra dichos índices de filas sobre el DATASET inicial 

```python
DATASET_AGRUPADO_STATE = DATASET.groupby(["State"])
DATASET_AGRUPADO_STATE
```

```python
DATASET_AGRUPADO_STATE.mean()
```

```python
#FILTRADO MEDIANTE LA CONDICIÓN (PROMEDIO DE Total day minutes > 199)
INDICES_FILAS = DATASET_AGRUPADO_STATE["Total day minutes"].filter(lambda X: X.mean() > 199).index
DATASET.loc[INDICES_FILAS , : ]
```
