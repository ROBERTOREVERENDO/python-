## PARTE 16.3: MODIFICACIÓN DE VALORES FALTANTES

```python
#CUANDO MODIFIQUES UN DATASET, CREA UN COPIA INDEPENDIENTE
DATASET_BK = DATASET.copy()
```

#### MODIFICACIÓN DE VALORES FALTANTES CON UN VALOR FIJO

```python
DATASET_BK = DATASET.copy()
DATASET_BK = DATASET_BK.fillna("-")
DATASET_BK.head()
```

#### MODIFICACIÓN DE VALORES FALTANTES CON UN VALOR FIJO EN 1 COLUMNA

```python
DATASET_BK = DATASET.copy()
DATASET_BK["Stars"] = DATASET_BK["Stars"].fillna("-")
DATASET_BK.head()
```

#### MODIFICACIÓN DE VALORES FALTANTES CON VALORES CALCULADOS

```python
DATASET_BK = DATASET.copy()
DATASET_BK["Stars"] = DATASET_BK["Stars"].fillna(DATASET_BK["Stars"].mean())
DATASET_BK.head()

#mean()
#median()
#min()
#max()
#quantile(0.5)
#quantile(0.75)
```

#### MODIFICACIÓN DE VALORES FALTANTES CON LÓGICAS BFILL Y FFILL

```python
DATASET_BK = DATASET.copy()
DATASET_BK["Stars"] = DATASET_BK["Stars"].fillna(method="backfill")
DATASET_BK.head()
```

```python
DATASET_BK = DATASET.copy()
DATASET_BK["Stars"] = DATASET_BK["Stars"].fillna(method="ffill")
DATASET_BK.head()
```

## PARTE 16.4: ELIMINACIÓN DE VALORES FALTANTES

* RECOMENDACIONES:
* 1) ELIMINA LAS FILAS Y COLUMNAS QUE NO TENGAN NINGÚN VALOR (TODOS SON VALORES FALTANTES)
* 2) ANALIZA SI LOS VALORES FALTANTES SE DEBEN ELIMINAR O MODIFICAR

#### PERSPECTIVA DE EJE

* 0 --> FILAS --> EJE HORIZONTAL
* 1 --> COLUMNAS --> EJE VERTICAL

#### PERSPECTIVA DEL CÓMO

* ANY --> ALGÚN VALOR FALTANTE
* ALL --> TODOS LOS VALORES FALTANTES

```python
DATASET_BK = DATASET.copy()
DATASET_BK = DATASET_BK.dropna(axis = 0, how = "any")
DATASET_BK.head()
```

```python
DATASET_BK = DATASET.copy()
DATASET_BK = DATASET_BK.dropna(axis = 0, how = "all")
DATASET_BK.head()
```

```python
DATASET_BK = DATASET.copy()
DATASET_BK = DATASET_BK.dropna(axis = 1, how = "any")
DATASET_BK.head()
```

```python
DATASET_BK = DATASET.copy()
DATASET_BK = DATASET_BK.dropna(axis = 1, how = "all")
DATASET_BK.head()
```
