## PARTE 20.1: DIVIDIR EL DATASET CON DISTRIBUCIÓN NORMAL

```python
import pandas as pd
RUTA = "DATASETS - CERTIFICACIÓN/17_1_VISUALIZACION_CHURN_MODEL.csv"
DATASET = pd.read_csv(RUTA)
DATASET.head()
```

```python
DATASET.shape[0]

print("80%-->", int(0.8 * DATASET.shape[0]))
print("20%-->", int(0.2 * DATASET.shape[0]))
```

```python
import numpy as np

#GENERAMOS UNA LISTA NORMAL QUE SE ASIGNARÁ A UNA COLUMNA DEL DATASET
LISTA_NORMAL = np.random.randn(DATASET.shape[0])
DATASET["NORMAL"] = LISTA_NORMAL

#CALCULAR EL PERCENTIL 80% (ES PARA SABER DÓNDE DIVIDIR)
LIMITE = np.quantile(LISTA_NORMAL,0.8)

#ANÁLISIS DE PERTENENCIA A DATASET ENTRENAMIENTO (80%) O PRUEBAS (20%)
LISTA_CONDICION = LISTA_NORMAL < LIMITE
DATASET["CONDICION"] = LISTA_CONDICION

#GENERAR DATASET ENTRENAMIENTO Y PRUEBAS
DATASET_ENTRENAMIENTO = DATASET[DATASET["CONDICION"] == True]
DATASET_PRUEBA = DATASET[DATASET["CONDICION"] == False]
```

## PARTE 20.2: DIVIDIR EL DATASET CON SHUFFLE

```python
import pandas as pd
RUTA = "DATASETS - CERTIFICACIÓN/17_1_VISUALIZACION_CHURN_MODEL.csv"
DATASET = pd.read_csv(RUTA)
DATASET.head()
```

```python
from sklearn.utils import shuffle

#REVOLOTEADO EL DATASET
DATASET = shuffle(DATASET)

#CALCULAMOS LA FILA DE CORTE
FILA_CORTE = int(DATASET.shape[0] * 0.8)

#GENERAR DATASET ENTRENAMIENTO Y PRUEBAS
DATASET_ENTRENAMIENTO = DATASET[:FILA_CORTE]
DATASET_PRUEBA = DATASET[FILA_CORTE:]
```
